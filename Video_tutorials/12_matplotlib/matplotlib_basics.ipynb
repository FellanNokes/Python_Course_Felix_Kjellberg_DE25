# matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-5,5)
x
```

```python
x.shape
```

matlab style interface

```python
f = lambda x: np.sin(x)
x = np.linspace(-5,5, 100)
plt.figure(dpi = 200)
plt.plot(x, f(x))
plt.xlabel("x")
plt.ylabel("y")
plt.title("$\sin{(x)}$")
```

oop approach

```python
fig, ax = plt.subplots(1)

ax.plot(x, f(x))
ax.set(xlabel="x", ylabel="y", title="$\sin{(x)}$")
```

## Subplots

```python
fig, ax = plt.subplots(1, 2, figsize=(16,4), dpi=200)

ax[0].plot(x, np.sin(x), label = "sin(x)")
ax[0].plot(x, np.cos(x), label = "cos(x)")
ax[0].legend()
ax[0].set(title="sin(x) and cos(x)", xlabel="x", ylabel="y")

ax[1].plot(x, np.cos(x))
ax[1].set(title="cos(x)", xlabel="x", ylabel="y")
```
